```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./input/NIceland_1Dall.dat", sep=' ')
df.head()
```

```python
z = df['z'] / 1000
mu = df['mu'] / (1e+9)
la = df['lambda'] / (1e+9)
rho = df['rho']
Qs = df['Qs']
Qp = df['Qp']
```

```python
plt.figure()
plt.plot(mu,z)
plt.xlabel('Rigidity [GPa]')
plt.ylabel('Depth [km]')
plt.hlines(y=-10, xmin=mu.min(), xmax=mu.max(), colors='k', linestyle='dashed')
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True)
fig.suptitle('Depth dependent parameters')

ax1.plot(rho, z)
ax1.set_ylabel('Depth [km]')
ax1.set_xlabel('Density [kg/m³]')
ax1.hlines(y=-10, xmin=rho.min(), xmax=rho.max(), colors='k', linestyle='dashed')

ax2.plot(mu, z, label='$\mu$')
ax2.plot(la, z, label='$\lambda$')
ax2.set_xlabel('Elastic Moduli [GPa]')
ax2.hlines(y=-10, xmin=la.min(), xmax=la.max(), colors='k', linestyle='dashed')
ax2.legend()

ax3.plot(Qp, z, label='Qp')
ax3.plot(Qs, z, label='Qs')
ax3.set_xlabel('Attenuation')
ax3.hlines(y=-10, xmin=Qs.min(), xmax=Qp.max(), colors='k', linestyle='dashed')
ax3.legend()

plt.tight_layout()
plt.savefig('./output/1Dstructure.png', dpi=300)
plt.show()
```
