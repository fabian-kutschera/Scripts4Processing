```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cE = pd.read_csv('output/MR_DR complex East_Mw6.68.csv')
cM = pd.read_csv('output/MR_DR complex Middle_Mw7.07.csv')
cW = pd.read_csv('output/MR_DR complex West_Mw6.74.csv')
```

```python
plt.figure()
plt.plot(cE['time'],cE['moment'], label='DR complex East')
plt.plot(cM['time'],cM['moment'], label='DR complex Middle')
plt.plot(cW['time'],cW['moment'], label='DR complex West')
plt.legend()
plt.grid()
plt.xlabel('Time [s]')
plt.ylabel('Moment Rate [Nm/s]')
plt.show()
```

```python
plt.figure()
plt.plot(cE['time'],cE['moment'], label='DR complex East')
plt.plot(cM['time'],cM['moment'], label='DR complex Middle')
plt.plot(cW['time'],cW['moment'], label='DR complex West')
plt.legend()
plt.grid()
plt.xlim(0,20)
plt.xlabel('Time [s]')
plt.ylabel('Moment Rate [Nm/s]')
plt.show()
```

```python
sE = pd.read_csv('output/MR_DR simple East_Mw7.34.csv')
sM = pd.read_csv('output/MR_DR simple Middle_Mw7.33.csv')
sW = pd.read_csv('output/MR_DR simple West_Mw7.34.csv')
```

```python
plt.figure()
plt.plot(sE['time'],sE['moment'], label='DR simple East')
plt.plot(sM['time'],sM['moment'], label='DR simple Middle')
plt.plot(sW['time'],sW['moment'], label='DR simple West')
plt.legend()
plt.grid()
plt.xlim(0,35)
plt.xlabel('Time [s]')
plt.ylabel('Moment Rate [Nm/s]')
plt.show()
```

```python
plt.rcParams["font.family"] = "serif"
plt.rc('xtick', labelsize=11) 
plt.rc('ytick', labelsize=11)

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, sharex=True, sharey=True)
fig.set_figwidth(14)
#fig.suptitle('Moment Rate', fontsize=14, y=1.01)

ax1.plot(sE['time'],sE['moment'], label='simple')
ax1.plot(cE['time'],cE['moment'], label='complex', linestyle='--', color='green')
ax1.grid()
ax1.set_xlabel('Time [s]', fontsize=12)
ax1.set_ylabel('Moment Rate [Nm/s]', fontsize=12)
ax1.set_title('DR East', fontsize=14)
ax1.legend()

ax2.plot(sM['time'],sM['moment'], label='simple')
ax2.plot(cM['time'],cM['moment'], label='complex', linestyle='--', color='green')
ax2.grid()
ax2.set_xlabel('Time [s]', fontsize=12)
ax2.set_title('DR Middle', fontsize=14)
ax2.legend()

ax3.plot(sW['time'],sW['moment'], label='simple')
ax3.plot(cW['time'],cW['moment'], label='complex', linestyle='--', color='green')
ax3.grid()
ax3.set_xlabel('Time [s]', fontsize=12)
ax3.set_title('DR West', fontsize=14)
ax3.legend()
ax3.set_xlim([0,40])

plt.savefig('MomentRate_subplot.png', dpi=300)
```
