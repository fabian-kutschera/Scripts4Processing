```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cE = pd.read_csv('MR DR complex East.csv')
cM = pd.read_csv('MR DR complex Middle.csv')
cW = pd.read_csv('MR DR complex West.csv')
```

```python
plt.figure()
plt.plot(cE['time'],cE['moment'], label='DR complex East')
plt.plot(cM['time'],cM['moment'], label='DR complex Middle')
plt.plot(cW['time'],cW['moment'], label='DR complex West')
plt.legend()
plt.grid()
plt.xlabel('Time [s]')
plt.ylabel('Moment Rate [Nm/s]')
plt.show()
```
