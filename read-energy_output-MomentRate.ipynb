```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams["font.family"] = "serif"
plt.rc('xtick', labelsize=11) 
plt.rc('ytick', labelsize=11)
```

```python
#link = "/import/freenas-m-05-seissol/kutschera/HIWI/fully-coupled/HFFZ/output_o6_II/HFFZ_fullycp_o6_newmaster-energy.csv"
#link = "/import/freenas-m-04-students/fkutschera/simulations/HFFZ_fullycp_o4_longrun/HFFZ_full-energy.csv"
link = "/import/freenas-m-04-students/fkutschera/simulations/HFFZ_fullycp_o4_180s/HFFZ_fullycp-energy.csv"
```

```python
df = pd.read_csv(link)
dt = df['time'][1]
moment_rate = np.gradient(df['seismic_moment'], dt)


fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(df['time'], moment_rate, label='Fully coupled simple East')


plt.legend()
plt.ylim(bottom=0)

#ax.spines['top'].set_visible(False)
#ax.spines['right'].set_visible(False)
#ax.get_xaxis().tick_bottom()
#ax.get_yaxis().tick_left()
ax.grid()

ax.set_ylabel('Moment Rate [Nm/s]')
ax.set_xlabel('Time [s]')
ax.set_xlim([0,35])

#plt.savefig('MRF.pdf', bbox_inches='tight')
plt.show()
```

```python
plt.plot(df['time'],df['seismic_moment'], label='Seismic moment')
plt.legend()
plt.grid()
plt.xlim(0,35)
M0 = df['seismic_moment'][len(df)-1]
Mw = 2/3 * (np.log10(M0) - 9.1)
print("M0 = {} and Mw = {}.".format(M0, Mw))
```

## theoretical upper limit of the tsunami energy derived by Bolshakova and Nosov (2011)

## earthquake energy

from (https://link.springer.com/article/10.1007/s00024-013-0730-6)

```python
E_TS_theoretical_max = 10 ** (2.0 * Mw - 1.7)
E_EQ = 10 ** (1.5 * Mw + 4.5)
print("E_TS_theoretical_max = {:.2} and E_EQ = {:.2}.".format(E_TS_theoretical_max, E_EQ))
```

## potential energy of initial elevation (tsunami energy) as a function of earthquake moment magnitude

```python
E_TS = 10 ** (2.33 * Mw - 5.59)
print("E_TS = {:.2}".format(E_TS))
```
