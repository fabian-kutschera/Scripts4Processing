```python
import pandas as pd
import numpy as np
```

```python
Xmin = -325e3; # converted to m
Xmax = 425e3;
Ymax = 375e3;
Ymin = -Ymax;
Zmin = -250e3;
Zmax = 2000; #height of the water layer

z_elastic = -0.1
z_acoustic = 1999.9

spacing = 50 # 50 m new from Carlos
```

```python
x_array = np.linspace(Xmin, Xmax, num=int((abs(Xmin)+Xmax)/spacing)+1)
y_array = np.linspace(Ymin, Ymax, num=int((abs(Ymin)+Ymax)/spacing)+1)
```

```python
print(len(x_array), len(y_array))
```

# Lines 1, 2, 3 - elastic 

```python
x_array
```

```python
y1 = -100e3
y2 = 0
y3 = 100e3
```

```python
# Line 1 - sub seafloor
df_line1_elastic = pd.DataFrame(columns=['x','y','z'])
df_line1_elastic["x"] = pd.DataFrame(x_array)
df_line1_elastic["y"] = y1
df_line1_elastic["z"] = z_elastic 
df_line1_elastic
```

```python
# Line 2 - sub seafloor
df_line2_elastic = pd.DataFrame(columns=['x','y','z'])
df_line2_elastic["x"] = pd.DataFrame(x_array)
df_line2_elastic["y"] = y2
df_line2_elastic["z"] = z_elastic 
df_line2_elastic
```

```python
# Line 3 - sub seafloor
df_line3_elastic = pd.DataFrame(columns=['x','y','z'])
df_line3_elastic["x"] = pd.DataFrame(x_array)
df_line3_elastic["y"] = y3
df_line3_elastic["z"] = z_elastic 
df_line3_elastic
```

# Lines 1, 2, 3 - acoustic

```python
# Line 1 - sub sea surface
df_line1_acoustic = pd.DataFrame(columns=['x','y','z'])
df_line1_acoustic["x"] = pd.DataFrame(x_array)
df_line1_acoustic["y"] = y1
df_line1_acoustic["z"] = z_acoustic
df_line1_acoustic
```

```python
# Line 2 - sub sea surface
df_line2_acoustic = pd.DataFrame(columns=['x','y','z'])
df_line2_acoustic["x"] = pd.DataFrame(x_array)
df_line2_acoustic["y"] = y2
df_line2_acoustic["z"] = z_acoustic
df_line2_acoustic
```

```python
# Line 3 - sub sea surface
df_line3_acoustic = pd.DataFrame(columns=['x','y','z'])
df_line3_acoustic["x"] = pd.DataFrame(x_array)
df_line3_acoustic["y"] = y3
df_line3_acoustic["z"] = z_acoustic
df_line3_acoustic
```

## Save receiver lines 1, 2, 3

```python
df_tot_123 = pd.concat([df_line1_elastic, df_line2_elastic, df_line3_elastic, df_line1_acoustic, df_line2_acoustic, df_line3_acoustic], ignore_index=True)
df_tot_123
```

```python
print(len(df_tot_123))
df_tot_123.to_csv("receiver_lines123.dat", index=False, header=False, sep=" ")
```
