```python
import pandas as pd
import numpy as np
```

```python
Xmin = -325e3; # converted to m
Xmax = 425e3;
Ymax = 375e3;
Ymin = -Ymax;
Zmin = -250e3;
Zmax = 2000; #height of the water layer

z_elastic = -0.1
z_acoustic = 1999.9

spacing_distance = 50 # 50 m new from Carlos
spacing_entries = 10 # 10 corresponds to 450m
```

```python
x_array = np.linspace(Xmin, Xmax, num=int((abs(Xmin)+Xmax)/spacing_distance)+1)
y_array = np.linspace(Ymin, Ymax, num=int((abs(Ymin)+Ymax)/spacing_distance)+1)
```

```python
print(len(x_array), len(y_array))
```

# Spacing of 50 m with inter-spacing of ~1 km

```python
# x-axis
b = x_array[::spacing_entries] # every 450 m

a = x_array[1:] # go to 2nd element -> 50 m
a = a[::spacing_entries]
```

```python
x_array = np.concatenate((a,b))
x_array.sort()
x_array = x_array[:-1] # delete last receiver because it has no partner :(
print(len(x_array))
print(x_array)
```

```python
# y-axis
b = y_array[::spacing_entries] # every 450 m

a = y_array[1:] # go to 2nd element -> 50 m
a = a[::spacing_entries]
```

```python
y_array = np.concatenate((a,b))
y_array.sort()
y_array = y_array[:-1] # delete last receiver because it has no partner :(
print(len(y_array))
print(y_array)
```
