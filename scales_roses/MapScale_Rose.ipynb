- Rose: https://docs.generic-mapping-tools.org/6.2/basemap.html?highlight=basemap#t
- Map scale: https://docs.generic-mapping-tools.org/6.2/basemap.html?highlight=basemap#l

```python
import pygmt
```

```python
fig = pygmt.Figure()
fig.basemap(region=[-6, 6, -6, 6], projection="M10c", frame=False, rose="JMC+w7c+f2+l")
# f = level of fanciness
fig.show()
fig.savefig("rose_fancy2.eps")
```

```python
fig = pygmt.Figure()
fig.basemap(region=[0, 0.5, 0, 1], projection="M10c", frame=False, rose="JMC+w10c+l")
fig.show()
fig.savefig("rose.eps")
```

```python
fig = pygmt.Figure()
pygmt.config(FONT=16)
fig.basemap(region=[-0, 1, 0, 1], projection="M10C", frame=False, map_scale="jBL+w50k+lkm+f") #+o0.5c/0.5c
fig.show()
fig.savefig("scale.eps", dpi=700)
```

```python
fig = pygmt.Figure()
pygmt.config(FONT=16)
fig.basemap(region=[-0, 1, 0, 1], projection="M12C", frame=False, map_scale="jBL+w24k+lDown-dip [km]") #+o0.5c/0.5c
fig.show()
fig.savefig("scale_Fra_width.eps", dpi=700)
```
